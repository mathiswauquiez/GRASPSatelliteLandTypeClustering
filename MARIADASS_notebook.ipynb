```python
import matplotlib.pyplot as plt

def visualize(feature):
  im = data[feature].to_numpy().reshape((len(data.index.levels[0]), len(data.index.levels[1])))
  plt.figure(figsize=(10, 8))
  plt.imshow(im, cmap='viridis', extent=[-180, 180, -90, 90])
  plt.colorbar(label=feature)
  plt.title(f'Visualization of {feature} Data')
  plt.xlabel('Longitude')
  plt.ylabel('Latitude')
  plt.grid(True)
  plt.show()
```

```python
DHR_variables = ['DHR1020', 'DHR443', 'DHR490', 'DHR565', 'DHR670', 'DHR865']
clustering_variables = DHR_variables + ['NDVI']

for variable in clustering_variables :
  visualize(variable)
```

# Clustering

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Variables DHR et NDVI pour le clustering
DHR_variables = ['DHR1020', 'DHR443', 'DHR490', 'DHR565', 'DHR670', 'DHR865']
clustering_variables = DHR_variables + ['NDVI']

# Sélection des données pour le clustering
clustering_data = data[clustering_variables].dropna()

# Création du modèle de clustering K-means
num_clusters = 10
kmeans = KMeans(n_clusters=num_clusters, random_state=42)

# Ajustement du modèle aux données
kmeans.fit(clustering_data)

# Ajout des étiquettes de cluster aux données utilisées pour le clustering
data['Cluster'] = np.nan
data.loc[clustering_data.index, 'Cluster'] = kmeans.labels_

# Définition d'une colormap pour les clusters
colors = plt.cm.tab10(np.linspace(0, 1, num_clusters))

# Fonction pour visualiser les clusters sur une carte
def visualize_clusters(cluster_feature):
    im = np.full((180, 360, 3), np.nan)  # Crée une matrice remplie de NaN pour stocker les clusters
    for index, row in data.iterrows():
        if not np.isnan(row['Cluster']):
            cluster_color = colors[int(row['Cluster'])][:3]  # Sélectionne la couleur correspondant au cluster et ignore la valeur alpha
            im[int(index[0] + 90), int(index[1] + 180)] = cluster_color  # Assigner la couleur du cluster à la position correspondante dans la matrice

    plt.figure(figsize=(10, 8))
    plt.imshow(im, extent=[-180, 180, -90, 90])
    plt.title('Visualization of Clusters')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.grid(True)
    plt.show()

# Visualisation des clusters sur une carte
visualize_clusters('Cluster')
```
